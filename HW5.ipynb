**Найти доверительный интервал для оценки математического ожидания $a$ нормального распределения с надежностью 0,95, зная выборочную среднюю $ \overline{x} = 75.12$, объем выборки $n=121$ и среднее квадратическое отклонение $\sigma=11$.**

Найдем доверительный интервал для математического ожидания a с
надежностью 0,95:

$$Ф(c_{\gamma}) = \frac{\gamma}{2} = 0.95/2 = 0.475$$


$$c_{\gamma} \approx 1.96$$

Получаем после подстановки известных данных: 

$$75.12 - 1.96 \cdot \frac{11}{\sqrt{121}} < a < 75.12 + 1.96 \cdot \frac{11}{\sqrt{121}}$$



$$73.16 < a < 77.08$$

**С целью размещения рекламы опрошено 600 телезрителей, из которых данную передачу смотрят 170 человек. С доверительной вероятностью 0,91 найти долю телезрителей, охваченных рекламой в наихудшем случае.**

$w = \frac{170}{600} \approx 0.2833$ - относительная частота события, $t = Ф^{-1} (\gamma/2) = Ф^{-1} (0.91/2) = 1.69$  (из таблицы функции Лапласа)

Подставляем данные:

$$0.2833 - 1.69\sqrt{\frac{0.2833(1-0.2833)}{600}} < p < 0.2833 + 1.69\sqrt{\frac{0.2833(1-0.2833)}{600}}$$

$$0.2522< p < 0.3143$$

Таким образом, получаем что худший вариант охвата телезрителей – 25,22% телезрителей.
